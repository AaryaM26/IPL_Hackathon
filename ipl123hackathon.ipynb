```python
#importing packages and data
import pandas as pd
match_data = pd.read_csv('/content/drive/MyDrive/all_matches.csv')
match_data.head()
```

```python
#checking whether data has null values or not
match_data.isnull().sum()
```

```python
#filling null values with 0
match_data.fillna(0)
```

```python
#first 6 over
match_data = match_data[match_data['ball'] < 6.0]
```

```python
#clubbing extra's and runs off bat
match_data['total_runs'] = match_data['runs_off_bat'] + match_data['extras']
```

```python
#removing the unnecessary data and grouping other attributes
match_data = match_data.groupby(['start_date','venue', 'innings', 'batting_team', 'bowling_team']).agg(total_runs = ('total_runs','sum'), wickets = ('wicket_type', 'count'),).reset_index()
match_data = match_data[match_data['innings'] < 3]
match_dataf=match_data.copy(deep=True);
match_data.head()
```

```python
#input team names
batting_team  = input("Enter batting team:")
bowling_team  = input("Enter bowling team:")
match_data = match_data[match_data['batting_team'] == batting_team]
match_data = match_data[match_data['bowling_team'] == bowling_team]
match_data.tail()
```

```python
#predicting score on basis of there past 5 matches
match_data1 = match_data.tail()
Total1 = match_data1['total_runs'].sum()
print(Total1/5)
```

```python
match_dataf.head()
```

```python
#predicting score on basis of stadium
stadium=input("Enter stadium:")
match_dataf = match_dataf[match_dataf['venue'] == stadium]
match_data2 = match_dataf.tail()

Total2 = match_data2['total_runs'].sum()
print(Total2/5)
```

```python
#averaging both
avg=(Total1 + Total2)/10
print(avg)
print('prediction = ',avg)
```

```python
#ball economy of each team
def ball_economy(bowling_team):
    teams2 ={'Royal Challengers Bangalore' : 51.14 ,'Chennai Super Kings' : 47.29, 'Rajasthan Royals':  57.81,'Kolkata Knight Riders': 58.14, 'Kings XI Punjab' : 50.7, 'Delhi Capitals' : 48.39,
    'Sunrisers Hyderabad':50.53 ,'Mumbai Indians':44.70}
    key=bowling_team
    if key in teams2.keys():
        
        print("balling economy average ", bowling_team, "is :", teams2[key])
        ball_avg=teams2[key]
        return (float (ball_avg))

balling_avg=ball_economy(bowling_team)
```
